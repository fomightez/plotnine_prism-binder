## ToothGrowth demo

This is the ToothGrowth demonstration from the main README. **Shows Plotnine base plot compared to the version with the plotnine-prism styling applied.**

Based on ['REAMDE' notebook from the plotnine-prism repo](https://nbviewer.org/github/pwwang/plotnine-prism/blob/master/examples/README.ipynb) and adapted to work in MyBinder-served sessions launched from [the repo plotnine_prism-binder](https://github.com/fomightez/plotnine_prism-binder).

```python
%%capture
!curl -OL https://raw.githubusercontent.com/pwwang/plotnine-prism/refs/heads/master/examples/nb_helpers.py #make sure current helper script present in session
from plotnine import *
from plotnine_prism import *

from nb_helpers import ToothGrowth
```

```python
base = (
    ggplot(ToothGrowth, aes(x = "dose", y = "len")) + 
    geom_violin(aes(colour = "dose", fill = "dose"), trim = False) + 
    geom_boxplot(aes(fill = "dose"), width = 0.2, colour = "black") 
)
```

```python
p1 = base + ylim(-5, 40)
p2 = (
    base + 
    scale_y_continuous(limits=[-15, 40], guide=guide_prism_offset_minor()) +
    scale_color_prism('floral') +
    scale_fill_prism('floral') +
    theme_prism()
)
print(p1, p2)
```
